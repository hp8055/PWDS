````python
#Q1
# Sure, here are five functions of the pandas library along with their execution:
 
# 1. **`read_csv()`**: Used to read data from a CSV file into a DataFrame.
 

import pandas as pd
   
# Reading a CSV file into a DataFrame
df = pd.read_csv('services.csv')

 
# 2. **`head()`**: Displays the first few rows of a DataFrame (default is the first 5 rows).
 
# Displaying the first 3 rows of the DataFrame
print(df.head(3))

 
# 3. **`info()`**: Provides a concise summary of the DataFrame, including the data types and number of non-null values in each column.
 
# Displaying information about the DataFrame
print(df.info())

 
# 4. **`describe()`**: Generates descriptive statistics of numerical columns in the DataFrame.
 

# Displaying summary statistics of numerical columns
print(df.describe())

 
# 5. **`value_counts()`**: Returns a Series containing counts of unique values in a column.
 

# Counting unique values in the 'City' column
locn_counts = df['location_id'].value_counts()
print(locn_counts)

 
# These functions are commonly used for data analysis and manipulation tasks in pandas.
````

````python
#Q2
def reindex_dataframe(df):
    # Resetting index and adding 1 to start from 1
    new_index = pd.Series(range(1, len(df) * 2, 2))
    # Assigning the new index to the DataFrame
    df = df.reset_index(drop=True)
    df.index = new_index
    return df
````

````python
d = {'a':[1,2,3,4],'b':[5,6,7,8],'c':[9,10,11,12]}
````

````python
df = pd.DataFrame(d)
````

````python
df = reindex_dataframe(df)
````

````python
#Q3

def sum3(df):
    x = 0
    y = 0 
    for items in df['values']:
        if y<3:
            x += items
            y +=1
    return x

    
    
    
````

````python
#Q9

d = {'a': [1, 2, 3, 4, 5],
     'values': [10, 20, 30, 40, 50],
     'b': [1, 2, 3, 4, 5]}
df = pd.DataFrame(d)

def new(column):
    return f"Mean is {column.mean()}, Median is {column.median()}, Standard deviation is {column.std()}"

result = new(df['values'])
print(result)
````
