```python
# Q2
# In Bokeh, glyphs are visual markers that represent data points on a plot.
# Glyphs can take various forms such as circles, lines, squares, and more, and they are used to visually encode data on a plot.
```

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers
output_file('test1.html')
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title="Example Glyphs Plot")
p.xaxis.axis_label = "X-Axis"
p.yaxis.axis_label = "Y-Axis"
p.circle(x, y, size=10, color="blue", alpha=0.5)
p.line(x, y, line_width=2, color="red", line_dash="dashed") 
show(p)
```

```python
# Q3
# You can customize the appearance of a Bokeh plot, including the axes, title, and legend, by using various properties and methods available in Bokeh's figure object. Here's a brief overview:

# Axes:

# Customize axis labels, ticks, tick labels, and grid lines using properties like x_axis_label, y_axis_label, xaxis.ticker, yaxis.ticker, xaxis.major_label_text_color, yaxis.major_label_text_font_size, etc.
# Title:

# Set the title of the plot using the title property of the figure object.
# Legend:

# Add a legend to the plot using the add_layout() method of the figure object and specifying a Legend object.
# Customize the legend appearance using properties like legend.label_text_font_size, legend.label_text_color, legend.location, etc.
```

```python
import bokeh.plotting as bp

# Prepare data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = bp.figure(title="Customized Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add glyphs to the figure
p.circle(x, y, size=10, color="blue", alpha=0.5, legend_label="Data Points")

# Customize axes
p.xaxis.major_label_text_font_size = "12pt"
p.yaxis.major_label_text_font_size = "12pt"
p.xaxis.axis_label_text_color = "green"
p.yaxis.axis_label_text_color = "green"

# Customize title
p.title.text_color = "blue"
p.title.text_font_size = "16pt"

# Customize legend
p.legend.title = "Legend"
p.legend.label_text_font_size = "10pt"
p.legend.label_text_color = "red"
p.legend.location = "top_right"

# Show the plot
bp.show(p)
```

```python
# Q4
# Bokeh server is a feature of the Bokeh library that allows you to create and serve interactive plots with real-time updates. With Bokeh server, you can build web applications that dynamically update plots based on user interactions or changes in the underlying data. 

# Here's a brief overview of how Bokeh server works:

# 1. **Define an Application**: You create a Python script that defines a Bokeh application by creating one or more Bokeh plots and specifying how they should update in response to user interactions or changes in data.

# 2. **Run the Bokeh Server**: You start the Bokeh server by running the Python script using the `bokeh serve` command. This command starts a web server that serves the Bokeh application.

# 3. **Access the Application**: Users can access the Bokeh application in a web browser by navigating to the URL provided by the Bokeh server.

# 4. **Interactive Updates**: Users can interact with the plots in the Bokeh application by panning, zooming, clicking, or dragging. The plots can update in real-time based on these interactions or changes in the underlying data.

# 5. **Server-side Callbacks**: You can define server-side callbacks in the Bokeh application that execute Python code in response to user interactions or changes in data. These callbacks can update the plots dynamically without requiring a full page reload.

# In summary, Bokeh server allows you to create interactive web applications with dynamic and real-time updating plots by serving Python-based Bokeh applications from a web server.
```
