```python
# Q1
# Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in Python.
# It provides a wide range of tools for creating various types of plots and charts, making it a powerful tool for data visualization.

# Five plots that can be plotted using the Pyplot module of Matplotlib include:

# Line plot: Visualizes data points connected by straight lines, commonly used to show trends or relationships over time.
# Scatter plot: Represents individual data points as markers on a plot, useful for visualizing relationships between two variables.
# Bar plot: Displays categorical data as rectangular bars, commonly used to compare the quantities of different categories.
# Histogram: Represents the distribution of numerical data by dividing the data into bins and counting the number of data points in each bin.
# Pie chart: Displays the proportions of different categories as slices of a circular pie, useful for showing the composition of a dataset.
```

```python
#Q2
# Scatter plot: Represents individual data points as markers on a plot, useful for visualizing relationships between two variables.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#Q3
# The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.
# It allows you to display multiple plots or charts together in a single visualization, making it easier to compare and analyze different datasets or aspects of your data. 
# Each subplot is specified by its position in the grid, determined by the number of rows and columns in the grid and the index of the subplot within the grid.
```

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(4, 1, 1) 
plt.plot(x1, y1)
plt.title('Plot 1')

plt.subplot(4, 1, 2) 
plt.plot(x2, y2)
plt.title('Plot 2')

plt.subplot(4, 1, 3) 
plt.plot(x3, y3)
plt.title('Plot 3')

plt.subplot(4, 1, 4) 
plt.plot(x4, y4)
plt.title('Plot 4')
plt.tight_layout()
```

```python
#Q4
# Bar plot: Displays categorical data as rectangular bars, commonly used to compare the quantities of different categories.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#Q5

# A box plot is a graphical representation that displays the distribution of a dataset based on five summary statistics: 
# minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum, allowing for the identification of outliers and overall data distribution.
# Box plots are used to:
# Summarize the distribution of a dataset and identify any outliers.
# Provide insights into the central tendency, spread, and skewness of the data.
# Compare distributions between different groups or categories in a dataset.
# Visualize the spread and variability of data in a concise and informative manner.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(x=box1)

```

```python
plt.boxplot(x=box2)
```
