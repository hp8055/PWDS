```python
#Q2
import seaborn as sns
ds = sns.load_dataset("fmri")
```

```python
sns.lineplot(x = ds.timepoint,y = ds.signal ,hue = ds.event ,style = ds.region ,markers = True)
```

```python
#Q3
ds1 = sns.load_dataset('titanic')
sns.boxplot(x=ds1.pclass,y= ds1.age )
```

```python
#Q4
ds2 = sns.load_dataset('diamonds')
sns.histplot(x=ds2.price,hue= ds2.cut,alpha = 0.75 )
```
